```python
def replace_typos(df):
    df.replace(r'\) ', ')', inplace=True, regex=True)
    df.replace(r'\( ', '(', inplace=True, regex=True)
    df.replace(r' \)', ')', inplace=True, regex=True)
    df.replace(r'[\w]\( ', r'[\w] \(', inplace=True, regex=True)
    df.replace([r'\(мс\)'], '(манчестер сити)', inplace=True, regex=True)
    df.replace([r'\(мю\)'], '(манчестер юнайтед)', inplace=True, regex=True)
    df.replace([r'\(реал\)'], '(реал мадрид)', inplace=True, regex=True)
    df.replace([r'\(бд\)', r'\(бвб\)'], '(боруссия дортмунд)', inplace=True, regex=True)
    df.replace(['хаффенхайм'], 'хоффенхайм', inplace=True, regex=True)
    df.replace(['ттх'], 'тоттенхэм', inplace=True, regex=True)
    df.replace(['холанн', 'холланд', 'халанд' ], 'холанд', inplace=True, regex=True)
    df.replace(['атм'], 'атлетико мадрид', inplace=True, regex=True)     
    df.replace(['марата'], 'мората', inplace=True, regex=True)
    df.replace(['мурьель', 'муриель'], 'муриэль', inplace=True, regex=True)
    df.replace(['марсиаль', 'марсияль'], 'марсьяль', inplace=True, regex=True)
    df.replace(['сорлот'], 'серлот', inplace=True, regex=True)
    df.replace(to_replace=['«', '»', '"'], value='', inplace=True, regex=True)
    df.replace(to_replace=['тоттэнхэм', 'тоттенхем'], value='тоттенхэм', inplace=True, regex=True)
    return df
```

```python
def transform_excel(matchday):
    """
    Трансформирует эксель из 3 колонок с бомебрами в одну
    """
    print(matchday)
    old_pd = pd.read_excel(f'data/bombs_{str(matchday)}.xlsx', index_col=0)
    raw_bombs = pd.DataFrame(pd.concat([old_pd.loc[:,'Бомбардир'],
                                        old_pd.loc[:,'Бомбардир.1'],
                                        old_pd.loc[:,'Бомбардир.2']], 
                                   axis=0),
                             columns=['Бомбардир'])
    raw_bombs.index.name = 'Name'
    raw_bombs.sort_values('Name', inplace=True)

        
    raw_bombs['Бомбардир'] = replace_typos(raw_bombs['Бомбардир'])
    
    writer = pd.ExcelWriter('data/bombs_'+str(matchday)+'.xlsx', engine='xlsxwriter')
    raw_bombs.to_excel(writer, sheet_name=str(matchday))
    writer.save()   
    
def process_raw_bombs(matchday):  
    raw_bombs = pd.read_excel(f'data/bombs_{str(matchday)}.xlsx', index_col=0)
    raw_bombs['Scorer'] = raw_bombs['Бомбардир'].apply(lambda x: x.split(' (')[0])
    raw_bombs['Team'] = raw_bombs['Бомбардир'].apply(lambda x: x.split(' (')[1][:-1])
    raw_bombs.drop('Бомбардир', axis=1, inplace=True)
    raw_bets = raw_bombs.copy()
    raw_bets['Goals'] = 0
    raw_bets['Points'] = 0
    raw_bets['Unique'] = 0
    raw_bombs = raw_bombs.drop_duplicates('Scorer')
    raw_bombs.set_index(['Scorer'], inplace=True)
    raw_bombs['Goals'] = 0
    raw_bombs['Picks'] = 0
    raw_bombs['Matchday'] = int(matchday)
    return raw_bombs, raw_bets
```

```python
# Transform all excel
for i in range(1, 10):
    transform_excel(i)
```

```python
def count_picks(bets, bombs):
    """
    Считает число пиков бомбера в туре
    """
    for scorer in list(bombs.index):
        picks = len(bets[bets['Scorer'] == scorer])
        bombs['Picks'].loc[scorer] = picks
        
def bomb_points(bets, bombs):
    """
    Начисляет очки
    """    
    scored = list(bombs[bombs['Goals'] > 0].index)

    for i in range(bets.shape[0]):
        scorer = bets['Scorer'].iloc[i]
        if scorer in scored:
            goals = bombs['Goals'].loc[scorer]
            bets.iloc[i, 2] = goals
            if len(bets[bets['Scorer'] == scorer]) == 1:
                points = (goals*2)+2 
                bets.iloc[i, 4] = 1
            else:
                points = goals*2
            bets.iloc[i, 3] = points
            #print(scorer, goals, points)
```
